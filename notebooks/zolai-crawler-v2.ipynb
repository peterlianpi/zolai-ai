# Zolai Dataset Crawler (v2)

**Purpose:** Scalable, session-resilient crawling for Zolai (Tedim) text data.

### Key Features:
1. **CPU Optimized:** Designed for unlimited Kaggle CPU hours.
2. **Session Resilience:** Saves progress to `crawler_state.json`. If Kaggle shuts down after 12h, simply restart to resume.
3. **JSONL Chunking:** Writes data in parts (`part_001.jsonl`, etc.) to prevent massive file corruption and high RAM usage.
4. **Smart Filtering:** Basic Zolai-text validation before saving.

```python
# 1. INSTALL & SETUP
import os
import json
import time
import hashlib
from pathlib import Path
from datetime import datetime
import httpx
from bs4 import BeautifulSoup
from tqdm.auto import tqdm

IS_KAGGLE = os.path.exists("/kaggle")
WORK_DIR = Path("/kaggle/working" if IS_KAGGLE else ".")
OUT_DIR = WORK_DIR / "zolai_crawled_parts"
OUT_DIR.mkdir(exist_ok=True)

STATE_FILE = WORK_DIR / "crawler_state.json"
LOG_FILE = WORK_DIR / "crawler.log"

def log(msg):
    ts = datetime.now().isoformat(timespec='seconds')
    line = f"[{ts}] {msg}"
    print(line)
    with open(LOG_FILE, "a", encoding="utf-8") as f:
        f.write(line + "\n")
```

```python
# 2. CONFIGURATION
MAX_LINES_PER_CHUNK = 50000  # Save a new file every N lines
REQUEST_TIMEOUT = 15.0
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"

# --- KAGGLE UPLOAD CONFIG ---
ENABLE_AUTO_UPLOAD = True
KAGGLE_DATASET_ID = "peterpausianlian/zolai-crawled-master"  # Change to your TARGET dataset slug

# Add your target URLs or sitemaps here
START_URLS = [
    # Example: "https://zolai-site.com/sitemap.xml"
]

def is_zolai_likely(text):
    """Simple heuristic: Check for common Zolai words like 'leh', 'hi', 'in'."""
    if not text or len(text) < 20: return False
    common = {" leh ", " hi ", " in ", " ka ", " na ", " a ", " uh "}
    t = text.lower()
    matches = sum(1 for w in common if w in t)
    return matches >= 2
```

```python
# 3. STATE MANAGEMENT & KAGGLE UPLOAD
import subprocess

def setup_kaggle_api():
    try:
        from kaggle_secrets import UserSecretsClient
        usc = UserSecretsClient()
        username = usc.get_secret(\"KAGGLE_USERNAME\")
        key = usc.get_secret(\"KAGGLE_KEY\")
        if not username or not key: return False
        kpath = Path.home() / \".kaggle\"
        kpath.mkdir(exist_ok=True)
        (kpath/\"kaggle.json\").write_text(json.dumps({\"username\": username, \"key\": key}))
        (kpath/\"kaggle.json\").chmod(0o600)
        return True
    except: return False

def push_to_kaggle(msg=\"Crawl update\"):
    if not ENABLE_AUTO_UPLOAD or not setup_kaggle_api(): return
    log(f\"Pushing to Kaggle: {KAGGLE_DATASET_ID}...\")
    meta = {\"id\": KAGGLE_DATASET_ID, \"licenses\": [{\"name\": \"CC0-1.0\"}]}
    with open(OUT_DIR / \"dataset-metadata.json\", \"w\") as f: json.dump(meta, f)
    subprocess.run([\"kaggle\", \"datasets\", \"version\", \"-p\", str(OUT_DIR), \"-m\", msg])

def load_state():
    if STATE_FILE.exists(): return json.loads(STATE_FILE.read_text())
    return {\"visited_urls\": [], \"pending_urls\": START_URLS, \"current_chunk_idx\": 1, \"lines_in_current_chunk\": 0}

def save_state(state):
    STATE_FILE.write_text(json.dumps(state, indent=2))
```

```python
# 4. CRAWLER CORE
def crawl(); push_to_kaggle(\"Final sync before session end\"):
    state = load_state()
    pending = set(state["pending_urls"])
    visited = set(state["visited_urls"])
    
    log(f"Starting crawl. Pending: {len(pending)}, Visited: {len(visited)}")
    
    client = httpx.Client(timeout=REQUEST_TIMEOUT, headers={"User-Agent": USER_AGENT}, follow_redirects=True)
    
    chunk_file = None
    
    try:
        while pending:
            url = pending.pop()
            if url in visited: continue
            
            try:
                resp = client.get(url)
                if resp.status_code != 200: continue
                
                soup = BeautifulSoup(resp.text, "html.parser")
                
                # --- EXTRACTION LOGIC (Custom per site) ---
                # This is a generic example extraction
                paragraphs = soup.find_all(['p', 'article'])
                text_data = []
                for p in paragraphs:
                    txt = p.get_text().strip()
                    if is_zolai_likely(txt):
                        text_data.append(txt)
                
                # --- SAVE LOGIC ---
                if text_data:
                    fname = OUT_DIR / f"part_{state['current_chunk_idx']:03d}.jsonl"
                    with open(fname, "a", encoding="utf-8") as f:
                        for t in text_data:
                            row = {"text": t, "source_url": url, "crawled_at": datetime.now().isoformat()}
                            f.write(json.dumps(row, ensure_ascii=False) + "\n")
                            state["lines_in_current_chunk"] += 1
                    
                    if state["lines_in_current_chunk"] >= MAX_LINES_PER_CHUNK:
                        log(f"Chunk {state['current_chunk_idx']} full. Moving to next.")
                        state["current_chunk_idx"] += 1
                        state["lines_in_current_chunk"] = 0
                
                # --- DISCOVERY LOGIC ---
                # Find new links on the same domain
                # for link in soup.find_all('a', href=True):
                #     new_url = link['href']... add to pending if relevant
                
                visited.add(url)
                
                # Periodic State Save
                if len(visited) % 10 == 0:
                    state["pending_urls"] = list(pending)
                    state["visited_urls"] = list(visited)
                    save_state(state)
                    
            except Exception as e:
                log(f"Error crawling {url}: {e}")
                continue
                
    finally:
        state["pending_urls"] = list(pending)
        state["visited_urls"] = list(visited)
        save_state(state)
        log("Crawl stopped. State saved.")

crawl(); push_to_kaggle(\"Final sync before session end\")
```
